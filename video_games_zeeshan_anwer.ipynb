```python
year_counts = df.Year_of_Release.value_counts()
year_order = year_counts.index
max_year_count = year_counts[year_counts.idxmax()]
max_year_prop = max_year_count/df.shape[0]

base_color = sb.color_palette()[0]
tick_props = np.arange(0, max_year_prop, 0.02)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]

plt.figure(figsize=(8,6))
sb.countplot(data = df, y = 'Year_of_Release', color = base_color, order = year_order)
plt.xticks(tick_props * df.shape[0], tick_names)
plt.xlabel('proportion');
plt.title('Year of Release Distribution', fontsize= 18);
```
